```
import nltk

def postag(book, sentence, target):
    word_list = book.split()
    postagged_word_list = nltk.pos_tag(word_list)
    list_of_tags = [i[1] for i in postagged_word_list]
    x_tagname = []
    y_tagcount = []
    for i in list(set(list_of_tags)):
      y_tagcount.append(list_of_tags.count(i))
      x_tagname.append(i)

    full_tags = pd.DataFrame({"tagname":x_tagname, "tagcount":y_tagcount})

    words_anne = []
    for i in range(len(target)):
      for e in range(len(target[i])):
    #      print(i, e)
         if target[i][e]==1:
            words_anne.extend(sentence[i][e].split())

    tag_anne = nltk.pos_tag(words_anne)

    list_of_tags_anne = [i[1] for i in tag_anne]
    x_tagname = []
    y_tagcount = []
    for i in list(set(list_of_tags_anne)):
      y_tagcount.append(list_of_tags_anne.count(i))
      x_tagname.append(i)

    # full_gables_tags.loc["anne_tagname"] = x_tagname
    # full_gables_tags.loc["anne_tagcount"] = y_tagcount
    full_tags["anne_tagcount"] = 0
    for i in range(len(full_tags)):
      if full_tags["tagname"][i] in x_tagname:
        full_tags["anne_tagcount"][i] = y_tagcount[x_tagname.index(full_tags["tagname"][i])]
    full_tags.loc[full_tags["tagcount"]==0, "tagcount"] = 0.0001
    full_tags.loc[full_tags["anne_tagcount"]==0, "anne_tagcount"] = 0.0001
    return full_tags
#     full_tags["anne_tagcount"] = full_tags["anne_tagcount"]

```
